```python
import time
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get("https://scholar.google.com/citations?view_op=view_org&hl=en&org=10241031385301082500")
```

```python
authorsDF = pd.DataFrame(columns=['user_id', 'name', 'affiliation'])
Paper = pd.DataFrame(columns=['title', 'authors', 'publication_data', 'description', 'cite_by'])
```

```python
dataPerpage = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div')
all_author = 1
all_paper = 1
```

```python
count = 0;
while len(dataPerpage) == 10:
    dataPerpage = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div')
    for i in range(len(dataPerpage)):
        dataPerpage = driver.find_elements_by_css_selector('#gsc_sa_ccl > div > div > div')

        a = dataPerpage[i].find_element_by_css_selector('a')
        user_id = a.get_attribute('href').split('=')[-1]
        name = dataPerpage[i].find_element_by_class_name('gs_ai_name').text
        affiliation = dataPerpage[i].find_element_by_class_name('gs_ai_aff').text

        authorsDF = authorsDF.append(
            {
                'user_id': user_id,
                'name': name,
                'affiliation': affiliation
            }, ignore_index=True
        )
        dataPerpage[i].find_element_by_css_selector('#gsc_sa_ccl > div > div > div > h3 > a').click()
        time.sleep(2)
        more_btn = driver.find_element_by_css_selector('#gsc_bpf_more')

        while more_btn.is_enabled():
            more_btn.click()
            time.sleep(1.5)
        time.sleep(3)
        j = 0
        for i in driver.find_elements_by_css_selector('#gsc_a_b > tr'):
            try:
                cite_by = i.find_element_by_css_selector('#gsc_a_b > tr > td.gsc_a_c > a').text
            except:
                cite_by = "-"

            i.find_element_by_css_selector('#gsc_a_b > tr > td.gsc_a_t > a').click()
            time.sleep(2)
            try:
                try:
                    title = driver.find_element_by_css_selector('#gsc_vcd_title > a').text
                except:
                    title = driver.find_element_by_css_selector('#gsc_vcd_title').text
            except:
                title = "-"

            try:
                authors = driver.find_element_by_css_selector(
                    '#gsc_vcd_table > div:nth-child(1) > div.gsc_vcd_value').text
            except:
                authors = "-"

            try:
                publication_data = driver.find_element_by_css_selector(
                    '#gsc_vcd_table > div:nth-child(2) > div.gsc_vcd_value').text
            except:
                publication_data = "-"

            try:
                try:
                    description = driver.find_element_by_css_selector('.gsh_csp').text
                except:
                    description = driver.find_element_by_css_selector('.gsh_small').text
            except:
                description = "-"

            Paper = Paper.append(
                {
                    'title': title,
                    'authors': authors,
                    'publication_data': publication_data,
                    'description': description,
                    'cite_by': cite_by
                }, ignore_index=True
            )
            print("คนที่ ", count + 1)
            print("user id ", user_id)
            print("Name ", name)
            print("Affiliation ", affiliation)
            print("งานที่ ", j + 1)
            print("ชื่อเรื่อง ",title)
            print("ผู้เขียน ",authors)
            print("วันที่ ",publication_data)
            print("อธิบาย ",description)
            print("cite",cite_by)
            print("คนทั้งหมด :",all_author," 'งานทั้งหมด ",all_paper)
            print("+++++++++++++++++++++++++++++++++++++++++")
            all_paper = all_paper + 1
            driver.find_element_by_css_selector('#gs_md_cita-d-x').click()
            j = j + 1
            # time.sleep(0.3)
        print("---------")
        driver.execute_script("window.history.go(-1)")
        all_author = all_author + 1
        count = count + 1
        authorsDF.to_csv("Author.csv")
        Paper.to_csv("Paper.csv")
        time.sleep(5)
    nextpage_btn = driver.find_element_by_css_selector('#gsc_authors_bottom_pag > div > button.gs_btnPR.gs_in_ib.gs_btn_half.gs_btn_lsb.gs_btn_srt.gsc_pgn_pnx')
    if nextpage_btn.is_enabled():
        nextpage_btn.click()
    time.sleep(2)
```
